```python
df['TransactionNo'] = df['TransactionNo'].astype('str')
```

```python
df['itemDescription'] = df['itemDescription'].str.strip()
```

```python
basket_Mirpur = (df[df['area'] =="Mirpur"]
          .groupby(['TransactionNo', 'itemDescription'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('TransactionNo'))

basket_Mohammadpur = (df[df['area'] =="Mohammadpur"]
          .groupby(['TransactionNo', 'itemDescription'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('TransactionNo'))

basket_Hazaribagh = (df[df['area'] =="Hazaribagh"]
          .groupby(['TransactionNo', 'itemDescription'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('TransactionNo'))

basket_Uttara = (df[df['area'] =="Uttara"]
          .groupby(['TransactionNo', 'itemDescription'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('TransactionNo'))
```

```python
# hot encode
def hot_encode(x):
    if(x<= 0):
        return 0
    if(x>= 1):
        return 1
```

```python
basket_encoded = basket_Mirpur.applymap(hot_encode)
basket_Mirpur = basket_encoded
```

```python
frq_items = apriori(basket_Mirpur, min_support = 0.001, use_colnames = True)
 

rules = association_rules(frq_items, metric ="lift", min_threshold = 1)
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False])

rules.head(20)
```
